```python
from sklearn import linear_model
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

Worldcup = {'score':[15,30,45,60,75,90],
            'soju':[1,3,4,7,7,11]}

Worldcup = pd.DataFrame(Worldcup)

import scipy.stats as stats

print(stats.pearsonr(Worldcup["score"], Worldcup["soju"]))

linear_regression = linear_model.LinearRegression()
linear_regression.fit(X = pd.DataFrame(Worldcup["score"]), y = Worldcup["soju"])

print('a value =', linear_regression.coef_)
print('b value = ', linear_regression.intercept_)

result = linear_regression.coef_*120+linear_regression.intercept_
print(result)
```
